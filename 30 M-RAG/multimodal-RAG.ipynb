```
# =====================================================================
# 1. 필수 패키지 설치 및 환경 세팅 (구글 코랩전용 라이브러리 포함)
# =====================================================================
!pip install -qU google-genai pillow numpy

import os
from google.colab import userdata
from google import genai
from google.genai import types
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# 코랩 Secrets에 등록한 'GEMINI_API_KEY' 자동 연동
os.environ["GEMINI_API_KEY"] = userdata.get('GEMINI_API_KEY')
client = genai.Client()

# =====================================================================
# 2. 멀티모달 데이터베이스(Knowledge Base) 생성
# =====================================================================
def create_solid_image(color_name, size=(300, 300)):
    return Image.new("RGB", size, color_name)

# 가상의 멀티모달 이미지 에셋 데이터베이스 구축
knowledge_base = [
    {
        "id": "asset_blue_cube",
        "text": "이 자산은 청량하고 시원한 느낌을 주는 바다 테마의 파란색 그래픽 요소입니다.",
        "image": create_solid_image("blue")
    },
    {
        "id": "asset_red_cube",
        "text": "이 자산은 정열적이고 강렬한 에너지를 표현하는 노을 테마의 빨간색 그래픽 요소입니다.",
        "image": create_solid_image("red")
    }
]

# 생성된 임시 이미지 시각화 확인
print(">> 생성된 데이터베이스 자산 확인:")
fig, axes = plt.subplots(1, 2, figsize=(6, 3))
for idx, item in enumerate(knowledge_base):
    axes[idx].imshow(item["image"])
    axes[idx].set_title(item["id"])
    axes[idx].axis('off')
plt.show()

# =====================================================================
# 3. 네이티브 멀티모달 임베딩 생성 (gemini-embedding-2)
# =====================================================================
print("\n>> 이미지 자산 벡터화 진행 중 (gemini-embedding-2)...")
for item in knowledge_base:
    response = client.models.embed_content(
        model="gemini-embedding-2",
        contents=item["image"]
    )
    item["vector"] = response.embeddings[0].values
print(f">> 완료! 각 이미지 자산이 {len(knowledge_base[0]['vector'])} 차원의 공통 공간 벡터로 정렬되었습니다.\n")

# =====================================================================
# 4. 유사도 검색(Retrieval) 및 최종 답변 생성(Generation)
# =====================================================================
def get_cosine_similarity(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))

# 사용자의 자연어 텍스트 질의
user_query = "여름 휴양지나 시원한 바다 파도 느낌에 어울리는 디자인 구성을 추천해 주세요."
print(f"[사용자 질문]: {user_query}\n")

# 질문 임베딩 추출 (텍스트 역시 동일한 임베딩 모델 공간 사용)
query_response = client.models.embed_content(
    model="gemini-embedding-2",
    contents=user_query
)
query_vector = query_response.embeddings[0].values

# 이미지 벡터 데이터셋과 코사인 유사도 매칭 수행 (Retrieval)
search_results = []
for item in knowledge_base:
    similarity = get_cosine_similarity(query_vector, item["vector"])
    search_results.append((similarity, item))

# 유사도가 가장 높은 순으로 정렬
search_results.sort(key=lambda x: x[0], reverse=True)
best_similarity, best_match_context = search_results[0]

print(f"[검색 완료] 가장 유사한 자산 ID: {best_match_context['id']} (유사도 점수: {best_similarity:.4f})")
print(">> Gemini가 추출된 멀티모달 컨텍스트를 해석하여 최종 답변 생성 중...\n")

# 매칭된 이미지와 문맥을 LLM에 전달하여 최종 가이드 답변 생성 (Generation)
final_response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=[
        "당신은 디자인 에셋 추천 RAG 어시스턴트입니다. 검색된 아래 [Context Data]의 이미지와 설명을 기반으로 사용자의 질문에 친절하게 답해 주세요.",
        "[Context Text]:", best_match_context["text"],
        "[Context Image]:", best_match_context["image"],
        "[User Query]:", user_query
    ]
)

print("[최종 생성 답변]:")
print(final_response.text)
```
