```python
import pandas as pd

# 파일의 공유 링크에서 ID 부분만 추출합니다.
file_id = '1HjG_3Yqlgih5zkQeFspkeACTXtdJx4OD'

# 구글 시트/드라이브 파일을 엑셀 형식으로 다운로드할 수 있는 URL로 변환합니다.
download_url = f'https://docs.google.com/spreadsheets/d/{file_id}/export?format=xlsx'

# pandas를 사용하여 엑셀 파일을 읽어옵니다.
try:
    df = pd.read_excel(download_url)

    # 데이터 확인 (상위 5개 행)
    print("파일 읽기 성공!")
    print(df.head())
except Exception as e:
    print(f"파일을 읽어오는 중 오류가 발생했습니다: {e}")
    print("참고: 파일의 공유 권한이 '링크가 있는 모든 사용자'로 설정되어 있는지 확인해주세요.")
```
