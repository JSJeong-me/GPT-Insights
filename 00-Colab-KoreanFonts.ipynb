```python
# ============================================================
# 1) 라이브러리 로드 및 Colab 업로드
# ============================================================
import io
import zipfile
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier

from imblearn.over_sampling import SMOTE

from docx import Document
from docx.shared import Inches
from google.colab import files

# 한글 폰트 설정 (설치된 Nanum Gothic 사용)
plt.rcParams["font.family"] = "NanumGothic"
plt.rcParams['axes.unicode_minus'] = False # To prevent the minus sign from appearing as a square


# ============================================================
# 2) CSV 업로드
# ============================================================
print("CSV 파일을 업로드하세요 (예: csi+wt_ver1_250312.csv)")
uploaded = files.upload()
csv_name = list(uploaded.keys())[0]
print("업로드 완료:", csv_name)

# ============================================================
# 3) 데이터 로드 & 기본 전처리
# ============================================================
def to_numeric_clean(s):
    return pd.to_numeric(pd.Series(s).astype(str).str.replace(r"[^0-9\.-]", "", regex=True), errors="coerce")

df = pd.read_csv(io.BytesIO(uploaded[csv_name]))
print("원본 shape:", df.shape)

# 필수 컬럼 존재 확인
needed_cols = [
    "사고원인", "일강수(mm)", "기온(℃)", "체감온도(℃)", "습도(%)",
    "공사종류", "공정률", "작업자수"
]
missing = [c for c in needed_cols if c not in df.columns]
if missing:
    raise ValueError(f"다음 필수 컬럼이 누락되었습니다: {missing}")

# 수치형 변환
df["일강수(mm)"] = to_numeric_clean(df["일강수(mm)"])
df["체감온도(℃)"] = to_numeric_clean(df["체감온도(℃)"])
# 기온/습도는 종종 이미 숫자지만, 안전하게 캐스팅
df["기온(℃)"] = to_numeric_clean(df["기온(℃)"])
df["습도(%)"] = to_numeric_clean(df["습도(%)"])

# 주요 분석용 서브셋
base_cols = ["사고원인", "일강수(mm)", "기온(℃)", "체감온도(℃)", "습도(%)"]
df_base = df[base_cols].dropna()

# ============================================================
# 4) 기술통계 & 사고원인별 기상조건 평균
# ============================================================
group_means = (
    df_base.groupby("사고원인")[["일강수(mm)", "기온(℃)", "체감온도(℃)", "습도(%)"]]
    .mean()
    .sort_index()
)
group_counts = df_base["사고원인"].value_counts()

print("\n[사고원인별 표본 수]")
print(group_counts.head(15))

print("\n[사고원인별 평균 기상 조건] (상위 10개 사고원인만 표시)")
print(group_means.loc[group_counts.head(10).index])

# ============================================================
# 5) 시각화 (상위 10개 사고원인 대상 Boxplot, Heatmap)
# ============================================================
top_causes = group_counts.head(10).index
df_plot = df_base[df_base["사고원인"].isin(top_causes)].copy()

# --- Boxplots (4개 변수) ---
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
for ax, col, title in zip(
    axes.ravel(),
    ["일강수(mm)", "기온(℃)", "체감온도(℃)", "습도(%)"],
    ["일강수(mm)", "기온(℃)", "체감온도(℃)", "습도(%)"],
):
    df_plot.boxplot(column=col, by="사고원인", ax=ax, rot=90)
    ax.set_title(f"사고원인별 {title} 분포")
    ax.set_xlabel("")
    ax.set_ylabel(title)
plt.suptitle("상위 10개 사고원인 - 기상 조건 분포(Boxplot)", y=1.02)
plt.tight_layout()
boxplot_path = "boxplots_top10.png"
plt.savefig(boxplot_path, dpi=150, bbox_inches="tight")
plt.close()

# --- Heatmap (사고원인별 평균) ---
pivot_means = (
    df_plot.groupby("사고원인")[["일강수(mm)", "기온(℃)", "체감온도(℃)", "습도(%)"]]
    .mean()
    .loc[top_causes]
)
plt.figure(figsize=(8, 6))
im = plt.imshow(pivot_means.values, aspect="auto")
plt.colorbar(im, fraction=0.046, pad=0.04)
plt.xticks(range(pivot_means.shape[1]), pivot_means.columns, rotation=45, ha="right")
plt.yticks(range(pivot_means.shape[0]), pivot_means.index)
plt.title("사고원인별 평균 기상 조건 (상위 10개)")
plt.tight_layout()
heatmap_path = "heatmap_means.png"
plt.savefig(heatmap_path, dpi=150, bbox_inches="tight")
plt.close()

# ============================================================
# 6) 예측모델 ①: 랜덤포레스트 (상위 10개 사고원인)
#     + 작업환경 변수(공사종류, 공정률, 작업자수) 포함
# ============================================================
df_model = df.copy()

# 범주형(구간) → Label Encoding
for col in ["공사종류", "공정률", "작업자수"]:
    df_model[col] = df_model[col].fillna("Unknown").astype(str)
    df_model[col] = LabelEncoder().fit_transform(df_model[col])

# 모델 입력 컬럼
model_cols = ["일강수(mm)", "기온(℃)", "체감온도(℃)", "습도(%)", "공사종류", "공정률", "작업자수"]

# 수치 변환 재적용
for c in ["일강수(mm)", "기온(℃)", "체감온도(℃)", "습도(%)"]:
    df_model[c] = to_numeric_clean(df_model[c])

df_model = df_model.dropna(subset=model_cols + ["사고원인"])

# 상위 10개 사고원인으로 제한
counts_all = df_model["사고원인"].value_counts()
top10 = counts_all.head(10).index
df_top = df_model[df_model["사고원인"].isin(top10)].copy()

# Encode target
le_y = LabelEncoder()
df_top["y"] = le_y.fit_transform(df_top["사고원인"])

X = df_top[model_cols].values
y = df_top["y"].values

# Train/Test split (Stratified)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# 모델 학습 (클래스 불균형 보정)
rf = RandomForestClassifier(
    n_estimators=300, random_state=42, class_weight="balanced"
)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

report_rf = classification_report(y_test, y_pred, target_names=le_y.classes_, zero_division=0)
print("\n[랜덤포레스트 성능(상위10, class_weight=balanced)]")
print(report_rf)

# 변수 중요도 (Baseline)
importances_base = pd.Series(rf.feature_importances_, index=model_cols).sort_values(ascending=False)

# 시각화: 변수 중요도, 혼동행렬
plt.figure(figsize=(7,5))
plt.barh(importances_base.index, importances_base.values)
plt.gca().invert_yaxis()
plt.title("변수 중요도 (RF, 상위10, class_weight=balanced)")
plt.tight_layout()
rf_varimp_path = "rf_varimp_base.png"
plt.savefig(rf_varimp_path, dpi=150, bbox_inches="tight")
plt.close()

cm_base = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8,6))
plt.imshow(cm_base, cmap="Blues")
plt.title("혼동행렬 (RF Base)")
plt.xlabel("예측")
plt.ylabel("실제")
plt.xticks(range(len(le_y.classes_)), le_y.classes_, rotation=45, ha="right")
plt.yticks(range(len(le_y.classes_)), le_y.classes_)
for (i, j), v in np.ndenumerate(cm_base):
    color = 'white' if cm_base[i, j] > cm_base.max() / 2 else 'black'
    # Highlight diagonal elements
    if i == j:
      plt.text(j, i, str(v), ha='center', va='center', color='red', fontweight='bold') # Highlight diagonal with red color
    else:
      plt.text(j, i, str(v), ha='center', va='center', color=color)
plt.tight_layout()
rf_cm_path = "rf_confusion_base.png"
plt.savefig(rf_cm_path, dpi=150, bbox_inches="tight")
plt.close()

# ============================================================
# 7) 예측모델 ②: SMOTE 적용 후 랜덤포레스트
# ============================================================
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(
    X_res, y_res, test_size=0.3, random_state=42, stratify=y_res
)

rf_smote = RandomForestClassifier(n_estimators=300, random_state=42)
rf_smote.fit(X_train_s, y_train_s)
y_pred_s = rf_smote.predict(X_test_s)

report_smote = classification_report(y_test_s, y_pred_s, target_names=le_y.classes_, zero_division=0)
print("\n[SMOTE 적용 후 랜덤포레스트 성능(상위10)]")
print(report_smote)

importances_smote = pd.Series(rf_smote.feature_importances_, index=model_cols).sort_values(ascending=False)

# 시각화: 변수 중요도, 혼동행렬 (SMOTE)
plt.figure(figsize=(7,5))
plt.barh(importances_smote.index, importances_smote.values)
plt.gca().invert_yaxis()
plt.title("변수 중요도 (RF + SMOTE, 상위10)")
plt.tight_layout()
rf_smote_varimp_path = "rf_varimp_smote.png"
plt.savefig(rf_smote_varimp_path, dpi=150, bbox_inches="tight")
plt.close()

cm_smote = confusion_matrix(y_test_s, y_pred_s)
plt.figure(figsize=(8,6))
plt.imshow(cm_smote, cmap="Blues")
plt.title("혼동행렬 (RF + SMOTE)")
plt.xlabel("예측")
plt.ylabel("실제")
plt.xticks(range(len(le_y.classes_)), le_y.classes_, rotation=45, ha="right")
plt.yticks(range(len(le_y.classes_)), le_y.classes_)
for (i, j), v in np.ndenumerate(cm_smote):
    color = 'white' if cm_smote[i, j] > cm_smote.max() / 2 else 'black'
    # Highlight diagonal elements
    if i == j:
      plt.text(j, i, str(v), ha='center', va='center', color='red', fontweight='bold') # Highlight diagonal with red color
    else:
      plt.text(j, i, str(v), ha='center', va='center', color=color)
plt.tight_layout()
rf_smote_cm_path = "rf_confusion_smote.png"
plt.savefig(rf_smote_cm_path, dpi=150, bbox_inches="tight")
plt.close()

# ============================================================
# 8) Word 보고서 생성 (분석 과정 + 결과 + 시각화 포함)
# ============================================================
doc = Document()
doc.add_heading("건설사고 정보 리포트 - 사고원인 예측 분석 보고서", 0)

doc.add_heading("1. 분석 목적", level=1)
doc.add_paragraph(
    "건설사고 데이터에서 기상조건(일강수, 기온, 체감온도, 습도)과 작업환경(공사종류, 공정률, 작업자수)을 바탕으로 "
    "사고원인과의 관계를 분석하고, 랜덤포레스트 기반 다중분류 모델로 예측 성능과 변수 중요도를 평가한다. "
    "데이터 불균형 해소를 위해 SMOTE 오버샘플링을 적용하여 소수 클래스의 예측 성능을 개선한다."
)

doc.add_heading("2. 데이터 전처리 및 기술통계", level=1)
doc.add_paragraph(
    "- 수치형 변환: 일강수, 기온, 체감온도, 습도\n"
    "- 범주형 인코딩: 공사종류, 공정률, 작업자수\n"
    "- 상위 빈도 사고원인 10개를 대상으로 상세 분석 수행"
)
doc.add_paragraph("[사고원인별 평균 기상 조건(상위10)] 표는 Colab 출력창을 참고.")

doc.add_heading("3. 시각화 (상위 10개 사고원인)", level=1)
doc.add_paragraph("3-1) 기상 조건 분포(Boxplot)")
doc.add_picture(boxplot_path, width=Inches(6))
doc.add_paragraph("3-2) 사고원인별 평균 기상 조건(Heatmap)")
doc.add_picture(heatmap_path, width=Inches(6))

doc.add_heading("4. 예측 모델 결과", level=1)
doc.add_paragraph("4-1) 랜덤포레스트(Base, class_weight=balanced):")
doc.add_paragraph(report_rf)
doc.add_paragraph("변수 중요도 (Base):")
doc.add_picture(rf_varimp_path, width=Inches(5.5))
doc.add_paragraph("혼동행렬 (Base):")
doc.add_picture(rf_cm_path, width=Inches(5.5))

doc.add_paragraph("4-2) 랜덤포레스트 + SMOTE:")
doc.add_paragraph(report_smote)
doc.add_paragraph("변수 중요도 (SMOTE):")
doc.add_picture(rf_smote_varimp_path, width=Inches(5.5))
doc.add_paragraph("혼동행렬 (SMOTE):")
doc.add_picture(rf_smote_cm_path, width=Inches(5.5))

doc.add_heading("5. 결론", level=1)
doc.add_paragraph(
    "기상조건(특히 습도, 강수량, 기온)과 작업환경(공사종류, 공정률, 작업자수)은 사고원인 예측에 유의한 기여를 한다. "
    "SMOTE 적용으로 소수 클래스의 재현율이 개선되며, 전반적인 성능 균형이 좋아진다. "
    "향후 작업유형, 안전관리 수준, 시간대 등 추가 변수 포함 시 설명력과 실용성이 더 향상될 수 있다."
)

docx_path = "건설사고_예측_분석_Report.docx"
doc.save(docx_path)

# ============================================================
# 9) 산출물 ZIP 묶기 + 자동 다운로드
# ============================================================
assets = [
    csv_name,
    boxplot_path,
    heatmap_path,
    rf_varimp_path,
    rf_cm_path,
    rf_smote_varimp_path,
    rf_smote_cm_path,
    docx_path
]
zip_name = "건설사고_예측_분석_산출물.zip"
with zipfile.ZipFile(zip_name, "w", zipfile.ZIP_DEFLATED) as zf:
    for p in assets:
        zf.write(p)

print("\n생성 파일 목록:")
for a in assets:
    print(" -", a)
print(" -", zip_name)

# 자동 다운로드
files.download(zip_name)
```
