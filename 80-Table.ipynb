```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
# Load the model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("microsoft/tapex-large-finetuned-wtq")
model = AutoModelForSeq2SeqLM.from_pretrained("microsoft/tapex-large-finetuned-wtq")
```

```python
# Example table data as a pandas DataFrame
data = {
    "Math": [10, 20],
    "English": [60, 70],
    # Add more columns and rows as needed
}
df = pd.DataFrame(data)

# Convert the DataFrame to a string in a tabular format
# table_string = df.to_markdown(index=False)

# Your query
query = "Waht is the average value of column name of Math?"
```

```python
import torch
```

```python
# Convert all data in DataFrame to string
df = df.applymap(str)

# Encode the table and query
inputs = tokenizer(table=df, queries=query, return_tensors="pt", truncation=True)

# Generate the answer
with torch.no_grad():
    outputs = model.generate(**inputs)

# Decode the generated answer
answer = tokenizer.decode(outputs[0], skip_special_tokens=True)

print(answer)
```
